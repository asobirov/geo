```python
# Cell 1 – setup & imports

import os
import requests
import zipfile
import io

# Base paths
BASE_DIR = 'data'
RAW_DIR  = os.path.join(BASE_DIR, 'raw')
PROC_DIR = os.path.join(BASE_DIR, 'processed')

# Ensure dirs exist
os.makedirs(RAW_DIR,  exist_ok=True)
os.makedirs(PROC_DIR, exist_ok=True)

print("✅ Directories ready:")
print("  raw →", RAW_DIR)
print("  processed →", PROC_DIR)
```

```python
DATASETS = {
    'gadm': [
        'https://geodata.ucdavis.edu/gadm/gadm4.1/shp/gadm41_UZB_shp.zip'
    ],
    'hdx': [
        'https://data.humdata.org/dataset/66f56d39-3f19-4798-a0df-a0ed9a4836ea/'
        'resource/eb1c227b-5b57-4008-be86-422f91edef6a/download/'
        'uzb_admbnda_adm0_2018b.zip',
        'https://data.humdata.org/dataset/66f56d39-3f19-4798-a0df-a0ed9a4836ea/'
        'resource/6a44247a-f4f5-421d-803a-893ecaf201e2/download/'
        'uzb_admbnda_adm1_2018b.zip',
        'https://data.humdata.org/dataset/66f56d39-3f19-4798-a0df-a0ed9a4836ea/'
        'resource/b9d8c3e7-b261-46a0-8ffc-b1c822c984c3/download/'
        'uzb_admbnda_adm2_2018b.zip'
    ],
    'geoBoundaries': [
        'https://github.com/wmgeolab/geoBoundaries/raw/main/'
        'sourceData/gbOpen/UZB_ADM0.zip',
        'https://github.com/wmgeolab/geoBoundaries/raw/main/'
        'sourceData/gbOpen/UZB_ADM1.zip',
        'https://github.com/wmgeolab/geoBoundaries/raw/main/'
        'sourceData/gbOpen/UZB_ADM2.zip'
    ]
}
```

```python
for dataset, urls in DATASETS.items():
    folder = os.path.join(RAW_DIR, dataset)
    os.makedirs(folder, exist_ok=True)
    for url in urls:
        fname = url.split('/')[-1]
        print(f"⏬ Downloading {fname} into {dataset}/ …")
        resp = requests.get(url)
        resp.raise_for_status()
        archive = zipfile.ZipFile(io.BytesIO(resp.content))
        print(f"   ↳ {len(archive.namelist())} files in archive")
        archive.extractall(folder)
        print(f"   ✔ Extracted to {folder}\n")
```

```python
# Convert raw shapefiles to GeoJSON, with logging and plots

import os
import logging
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. Logging setup
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s %(levelname)s %(message)s"
)
logger = logging.getLogger("admin-boundaries")

# 2. Paths
RAW_BASE  = "data/uzbekistan/admin-boundaries/raw"
PROC_BASE = "data/uzbekistan/admin-boundaries/processed"

# 3. Dataset configs: folder name → filename pattern
SOURCES = {
    "gadm":       "gadm41_UZB_{l}.shp",            # from GADM v4.1
    "hdx":        "uzb_admbnda_adm{l}_2018b.shp",  # from HDX COD-AB
    "geoBoundaries":"UZB_ADM{l}.shp"               # from geoBoundaries
}

for src, pattern in SOURCES.items():
    src_raw = os.path.join(RAW_BASE, src)
    for lvl in (0, 1, 2):
        shp = os.path.join(src_raw, pattern.format(l=lvl))
        if not os.path.exists(shp):
            logger.warning(f"Missing {src} ADM{lvl} shapefile: {shp}")
            continue

        logger.info(f"Reading {src} ADM{lvl} → {shp}")
        gdf = gpd.read_file(shp)

        # Reproject to WGS84 lon/lat (EPSG:4326)
        gdf = gdf.to_crs(epsg=4326)
        out_fn = f"{src}_adm{lvl}.geojson"
        out_path = os.path.join(PROC_BASE, out_fn)

        # Write GeoJSON
        gdf.to_file(out_path, driver="GeoJSON")
        logger.info(f"  → Wrote {len(gdf)} features to {out_path}")

        # Quick boundary plot
        fig, ax = plt.subplots(figsize=(6, 4))
        gdf.boundary.plot(ax=ax, linewidth=0.8)
        ax.set_title(f"{src} ADM{lvl} boundaries ({len(gdf)} features)")
        ax.set_xlabel("Longitude")
        ax.set_ylabel("Latitude")
        plt.tight_layout()
        plt.show()
```
